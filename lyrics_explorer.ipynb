# English Song Database Lyrics Explorer

This notebook provides a simple way to access and view lyrics from the `English_Song_Database`.

```python
import pandas as pd
import os

# Define the path to the CSV directory
dataset_path = 'English_Song_Database/csv'

# Check if directory exists
if not os.path.exists(dataset_path):
    print(f"Error: Directory '{dataset_path}' not found. Please ensure the path is correct.")
else:
    print(f"Found database at: {dataset_path}")
```

### 1. List Available Artists
Run the cell below to see which artists are available in the database.

```python
# Get all CSV files (Artists)
files = [f for f in os.listdir(dataset_path) if f.endswith('.csv')]
artists = [f.replace('.csv', '') for f in files]

print("Available Artists:")
for i, artist in enumerate(artists):
    print(f"{i + 1}. {artist}")
```

### 2. Select an Artist
Change the `artist_name` variable below to one of the artist names from the list above (e.g., `'ArianaGrande'`).

```python
artist_name = 'ArianaGrande'  # <--- CHANGE THIS to your desired artist

file_path = os.path.join(dataset_path, f"{artist_name}.csv")

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"\nSuccessfully loaded {len(df)} songs for {artist_name}.")
    display(df.head()) # Show first few rows
else:
    print(f"Error: File for {artist_name} not found.")
```

### 3. List All Songs by this Artist

```python
print(f"Songs by {artist_name}:\n")
for title in df['Title'].unique():
    print(f"- {title}")
```

### 4. Get Lyrics for a Specific Song
Copy a song title from the list above and paste it into the `song_title` variable.

```python
song_title = '7 rings' # <--- CHANGE THIS to the song you want

# Find the song in the dataframe
song_row = df[df['Title'] == song_title]

if not song_row.empty:
    lyrics = song_row.iloc[0]['Lyric']
    title = song_row.iloc[0]['Title']
    album = song_row.iloc[0]['Album']
    
    print(f"--- {title} ---")
    print(f"Album: {album}\n")
    print(lyrics)
else:
    print(f"Song '{song_title}' not found. Please check the spelling.")
```
